## Assignment

- Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, men and women (remember to use the exact column titles)

```python
column = ['Sample_size', 'Median', 'Employed', 'Full_time','Unemplyment_rate', 'Men', 'Women']
obj = plt.figure(figsize=(5,12))
for i in range(1,5):
    ax = obj.add_subplot(4,1,i)
    ax = majors_df[column[i-1]].plot(kind ='hist')
```
